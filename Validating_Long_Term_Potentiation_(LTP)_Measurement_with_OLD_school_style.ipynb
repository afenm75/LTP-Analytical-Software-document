**Developing potentiation measurment model based on Old school study:**



***op1_data_extraction.py***
**Objective**: Extracts electrophysiology data from msrd files, including waveforms, meta-information, and timestamps, using NeuroExplorer.

**Output**: Organized data in structured formats like dataframes for further analysis.


***op2_waveforms_noise_preparation-per electrode:***

**Objective:** To process LTP waveforms by performing downsampling, bandpass filtering, and smoothing. Common Average Referencing (CAR) will be employed for denoising, with TTX applied for each electrode. While artifact detection is tested, artifact removal will primarily be conducted manually on the waveforms.
image.png



***op3_adding_meta_info***

**Objective:** To integrate clinical characteristics of patients and electrode-specific data into the LTP waveforms. Focus on analyzing the last 5 minutes of TTX application, completely excluding the periods used for stimulation and inhibitor application. After this step, the most responsive signals will be prepared and ready for further analysis.

***op4_potentiation_measurment***

**Objective:** To average the most responsive waveforms before and after stimulation for each electrode. The enhancement in potentiation will be measured using the minimum peak, which is selected based on findings from our previous study. In that project, we tracked how potentiation changes for all samples and each group of patients with different early AD pathology, as outlined in the following analysis.
